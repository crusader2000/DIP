```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.express as px
import os
import time
IMAGES_FOLDER_PATH = "../images"
```

```python
coins = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'coins.jpg'),0)
# plt.imshow(thumb,cmap='gray')
# plt.show()

ret,coins = cv2.threshold(coins,132,255,cv2.THRESH_BINARY)
plt.figure(figsize=(12,10))
plt.imshow(coins,cmap='gray')
plt.show()
coins = np.int64(coins/255)
```

## Part 1

```python
visited = np.zeros(coins.shape)
labels = np.zeros(coins.shape)

import sys
sys.setrecursionlimit(10000000)
def dfs(img,i,j,color):
    if i<0:
        return
    if j<0:
        return
    if i>=img.shape[0]:
        return
    if j>=img.shape[1]:
        return
    if visited[i,j]:
        return
    if img[i,j]==0:
        return
    
    labels[i,j] = color
    visited[i,j] = 1
    
    dfs(img,i-1,j-1,color)
    dfs(img,i-1,j,color)
    dfs(img,i-1,j+1,color)
    dfs(img,i+1,j-1,color)
    dfs(img,i+1,j+1,color)
    dfs(img,i+1,j,color)
    dfs(img,i,j+1,color)
    dfs(img,i,j,color)
    dfs(img,i,j-1,color)
    


def get_boundary_coins(img):
    color = 1

    h,w = img.shape
    for i in [0,h-1]:
        for j in range(w):
            if img[i,j]==1:
                dfs(img,i,j,color)
    
    for i in range(h):
        for j in [0,w-1]:
            if img[i,j]==1:
                dfs(img,i,j,color)
 
 
    return labels
```

```python
boundary_coins = get_boundary_coins(coins)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(boundary_coins,cmap='gray')
plt.show()
```

## Part 2 & Part 3

```python
visited = np.zeros(coins.shape)
labels = np.zeros(coins.shape)

def dfs(img,i,j,color):
    if i<0:
        return 0
    if j<0:
        return 0
    if i>=img.shape[0]:
        return 0
    if j>=img.shape[1]:
        return 0
    if visited[i,j]:
        return 0
    if img[i,j]==0:
        return 0
    
    labels[i,j] = color
    visited[i,j] = 1
    
    return (1 + dfs(img,i-1,j-1,color) + dfs(img,i-1,j,color) + dfs(img,i-1,j+1,color) + dfs(img,i+1,j-1,color) +  dfs(img,i+1,j+1,color) + dfs(img,i+1,j,color) + dfs(img,i,j+1,color) + dfs(img,i,j,color) + dfs(img,i,j-1,color))


def connected_components(img):
    color = 1
    threshold = 580
    
    non_overlap_colors = []
    overlap_colors = []
    
    h,w = img.shape
    for i in range(h):
        for j in range(w):
            if visited[i,j]==0 and img[i,j]==1:
                area_covered = dfs(img,i,j,color)
                if area_covered > threshold:
                    overlap_colors.append(color)
                else:
                    non_overlap_colors.append(color)
                color = color + 1
    
    overlap_img = np.zeros(img.shape)
    non_overlap_img = np.zeros(img.shape)
    for i in range(h):
        for j in range(w):
            if labels[i,j]:
                if labels[i,j] in non_overlap_colors:
                    non_overlap_img[i,j] = 1
                else:
                    overlap_img[i,j] = 1
 
    return overlap_img,non_overlap_img
```

```python
overlap_img,non_overlap_img = connected_components(coins)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(overlap_img,cmap='gray')
plt.title('Overlapping Coins')
plt.show()
plt.figure(figsize=(12,10))
plt.imshow(non_overlap_img,cmap='gray')
plt.title('Non Overlapping Coins')
plt.show()
```
