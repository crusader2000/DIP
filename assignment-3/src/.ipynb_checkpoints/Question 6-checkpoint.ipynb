```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
from skimage.color import label2rgb
# import plotly.express as px
import os
import time
IMAGES_FOLDER_PATH = "../images"
```

```python
objects = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'objects.png'),0)
# plt.imshow(thumb,cmap='gray')
# plt.show()

ret,objects = cv2.threshold(objects,132,255,cv2.THRESH_BINARY)
plt.figure(figsize=(12,10))
plt.imshow(objects,cmap='gray')
plt.show()
objects = np.int64(objects/255)
```

## Part 1

```python
def ccl_2_pass(img):
    h,w = img.shape
    labels = np.int64(np.zeros(img.shape))

    assoc = {}
    clr = 1
    if img[0,0] == 1:
        labels[0,0] = 1
        assoc[1] = -1
        clr = 2
            
    for j in range(1,w):
        if img[0,j] == 1:
            if img[0,j-1] == 1:
                labels[0,j] = labels[0,j-1]
            else:
                labels[0,j] = clr
                assoc[clr] = -1
                clr = clr + 1
                
    for i in range(1,h):
        for j in range(w):
            if img[i,j] == 1:
                prev_labels = []
                
                if labels[i-1,j]>0:
                    prev_labels.append(labels[i-1,j])
                
                if j > 1 :
                    if labels[i-1,j-1]>0:
                        prev_labels.append(labels[i-1,j-1])
                    if labels[i,j-1]>0:
                        prev_labels.append(labels[i,j-1])
                if (j+1) < w:
                    if labels[i-1,j+1]>0:
                        prev_labels.append(labels[i-1,j+1])

                if len(prev_labels) > 0:
                    prev_labels = np.array(prev_labels)
                    labels[i,j] = np.min(prev_labels)
                    for x in prev_labels:
                        if x != labels[i,j]:
                            assoc[x] = labels[i,j]
                else:
                    labels[i,j] = clr
                    assoc[clr] = -1
                    clr = clr +1
    for k,v in assoc.items():
        temp = v

        while(v!=-1 and assoc[temp] != -1):
            temp = assoc[temp]
        assoc[k] = temp
    
    for i in range(h):
        for j in range(w):
            if labels[i,j] and assoc[labels[i,j]]!=-1:
                labels[i,j] = assoc[labels[i,j]]

    return labels
```

```python
labels = ccl_2_pass(objects)
```

```python
label_map = label2rgb(labels,bg_label=0)
```

```python
plt.imshow(label_map)
plt.show()
print(label_map.shape)
```

## Part 1

We can scale the labelmap using a transformation matrix.
$$ T = \begin{bmatrix}
k & 0 \\
0 & k 
\end{bmatrix}$$

The Transformation is 
$ \begin{bmatrix}
X \\
Y \\ 
\end{bmatrix}
$
=$\begin{bmatrix}
k & 0 \\
0 & k 
\end{bmatrix}$$\begin{bmatrix}
x  \\
y  
\end{bmatrix}$

where $X,Y$ are new co-ordinates and $x,y$ are original co-ordinates. Using this formula directly will create gaps and holes in the final image. Therefore reverse mapping is used. 

After reverse mapping, interpolation schemes are used to find the intensity for the new co-ordinate

The code below uses nearest neighbour to find the intensity values


```python
def upscale(img,k):
    mat_inv = np.array([[1/k,0],[0,1/k]])
    h,w = img.shape
    
    output = np.zeros((h*k,w*k))
    h1,w1 = output.shape
    for i in range(h1):
        for j in range(w1):
            x,y = np.dot(mat_inv,np.array([[i],[j]]))
            x = int(x[0])
            y = int(y[0])
            if x>=0 and x<h and y>=0 and y<w:
                output[i,j] = img[x,y]
    return output
```

```python
upscaled = upscale(labels,5)
```

```python
label_map_upscaled = label2rgb(upscaled,bg_label=0)
plt.imshow(label_map)
plt.show()
print(label_map.shape)
```

## Part 2

We can scale the labelmap using a transformation matrix.
$$ T = \begin{bmatrix}
cos(\theta) & -sin(\theta) \\
sin(\theta) & cos(\theta)
\end{bmatrix}$$

The Transformation is 
$ \begin{bmatrix}
X \\
Y \\ 
\end{bmatrix}
$
=$\begin{bmatrix}
cos(\theta) & -sin(\theta) \\
sin(\theta) & cos(\theta)
\end{bmatrix}$$\begin{bmatrix}
x  \\
y  
\end{bmatrix}$

where $X,Y$ are new co-ordinates and $x,y$ are original co-ordinates. Using this formula directly will create gaps and holes in the final image. Therefore reverse mapping is used. 

After reverse mapping, interpolation schemes are used to find the intensity for the new co-ordinate


The code below uses nearest neighbour to find the intensity values


```python
def rotate(img,alpha,retain = True):
    
    alpha = np.radians(alpha)

    h, w,c = img.shape
    cy, cx = h / 2, w / 2

    a = np.round(np.math.cos(alpha),2)
    b = np.round(np.math.sin(alpha),2)
    xc1 = -cx
    xc2 = cx
    yc1 = -cy
    yc2 = cy

    rot_matrix = np.array([
    [   
        a, -b, xc1 * a - yc1 * b + xc2],[ b, a, xc1 * b + yc1 * a + yc2],[0,0,1]
    ])
    
    rot_matrix_inv = np.linalg.inv(rot_matrix)
    output = np.zeros((h,w,3))
    for i in range(h):
        for j in range(w):
            y,x,_ = np.dot(rot_matrix_inv,np.array([[j],[i],[1]]))
            x = int(np.rint(x[0]))
            y = int(np.rint(y[0]))
            if x>=0 and x<h and y>=0 and y<w:
                output[i,j][0] = img[x,y][0]
                output[i,j][1] = img[x,y][1]
                output[i,j][2] = img[x,y][2]

    if not retain:
        new_h_min = 100000000000000
        new_w_min = 100000000000000
        new_h_max = -100000000000000
        new_w_max = -100000000000000
        for i in range(h):
            for j in range(w):
                rotate_pts = rot_matrix @ np.array([[j],[i],[1]])
                new_h_min = min(new_h_min,int(rotate_pts[1]))  
                new_w_min = min(new_w_min,int(rotate_pts[0]))  
                new_h_max = max(new_h_max,int(rotate_pts[1]))  
                new_w_max = max(new_w_max,int(rotate_pts[0]))   

        new_h = new_h_max - new_h_min
        new_w = new_w_max - new_w_min

        rot_matrix2 = np.array([
        [   
            a, -b, xc1 * a - yc1 * b + new_w/2],[ b, a, xc1 * b + yc1 * a + new_h/2],[0,0,1]
        ])

        new_h_min = 100000000000000
        new_w_min = 100000000000000
        new_h_max = -100000000000000
        new_w_max = -100000000000000
        for i in range(h):
            for j in range(w):
                rotate_pts = rot_matrix2 @ np.array([[j],[i],[1]])
                new_h_min = min(new_h_min,int(rotate_pts[1]))  
                new_w_min = min(new_w_min,int(rotate_pts[0]))  
                new_h_max = max(new_h_max,int(rotate_pts[1]))  
                new_w_max = max(new_w_max,int(rotate_pts[0]))   

        new_h = new_h_max - new_h_min
        new_w = new_w_max - new_w_min

        output = np.zeros((new_h,new_w,3))

        rot_matrix2_inv = np.linalg.inv(rot_matrix2)


        for i in range(new_h):
            for j in range(new_w):
                y,x,_ = np.dot(rot_matrix2_inv,np.array([[j],[i],[1]]))
                x = int(np.rint(x[0]))
                y = int(np.rint(y[0]))
                if x>=0 and x<h and y>=0 and y<w:
                    output[i,j][0] = img[x,y][0]
                    output[i,j][1] = img[x,y][1]
                    output[i,j][2] = img[x,y][2]

    return output
```

```python
output_crop = rotate(label_map,30,retain = True)
output = rotate(label_map,30,retain = False)
```

```python
plt.figure(figsize=(15,15))
plt.subplot(121)
plt.imshow(output_crop,cmap='gray')
plt.title("Rotate Angel = 30 (Retain = True)")
plt.subplot(122)
plt.imshow(output,cmap='gray')
plt.title("Rotate Angel = 30 (Retain = False)")
plt.show()
```
