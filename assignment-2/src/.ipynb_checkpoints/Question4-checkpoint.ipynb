```python
def bilateral_filter(img,k,sigma_s,sigma_v):

    h,w,_ = img.shape
    
    output = np.zeros([h,w,3])
    
    padding_img = np.zeros([h+k-1,w+k-1,3],dtype='int64')
    padding_img[int((k-1)/2):int(h+(k-1)/2),int((k-1)/2):int(w+(k-1)/2),0:3] = img
    spatial_kernel = np.zeros([k,k])
    
    center = int((k-1)/2)
    for i in range(k):
        for j in range(k):
            spatial_kernel[i][j] = np.exp(-(((i - center)**2 + (j-center)**2) / (2 * sigma_s * sigma_s)))
    
    for i in range(int((k-1)/2),h+int((k-1)/2)):
        for j in range(int((k-1)/2),w+int((k-1)/2)):     
            for layer in range(3):
                area_obs = padding_img[i-int((k-1)/2):i+int((k-1)/2)+1,j-int((k-1)/2):j+int((k-1)/2)+1,layer:layer+1].reshape(k,k)                
                value_kernel = padding_img[i-int((k-1)/2):i+int((k-1)/2)+1,j-int((k-1)/2):j+int((k-1)/2)+1,layer:layer+1].reshape(k,k) - padding_img[i][j][layer]
                value_kernel = np.exp(-(value_kernel*value_kernel)/(2*sigma_v*sigma_v))
                
                denominator = spatial_kernel * value_kernel
                numerator = area_obs*denominator

                output[i-int((k-1)/2)][j-int((k-1)/2)][layer] = int(sum(sum(numerator)) / sum(sum(denominator)))

    return np.int64(output)
```

```python
filtered = bilateral_filter(car,9,200,50)
plt.imshow(filtered)
plt.show()
```

```python
def blurring_gaussian(img,k,sigma):
    h,w = img.shape
    
    output = np.zeros([h,w])
    
    padding_img = np.zeros([h+k-1,w+k-1],dtype='int64')
    padding_img[int((k-1)/2):int(h+(k-1)/2),int((k-1)/2):int(w+(k-1)/2)] = img
    
    kernel = np.zeros([k,k])
    
    center = int((k-1)/2)
    for i in range(k):
        for j in range(k):
            kernel[i][j] = np.exp(-(((i - center)**2 + (j-center)**2) / (2 * sigma * sigma)))
    
    kernel = kernel/sum(sum(kernel))
    
    for i in range(int((k-1)/2),h+int((k-1)/2)):
        for j in range(int((k-1)/2),w+int((k-1)/2)):     
            area_obs = padding_img[i-int((k-1)/2):i+int((k-1)/2)+1,j-int((k-1)/2):j+int((k-1)/2)+1].reshape(k,k)  
            output[i-int((k-1)/2)][j-int((k-1)/2)] = int(sum(sum(kernel*area_obs)))
    
    return output

def blurring_mean(img,k):
    h,w = img.shape
    
    output = np.zeros([h,w])
    
    padding_img = np.zeros([h+k-1,w+k-1],dtype='int64')
    padding_img[int((k-1)/2):int(h+(k-1)/2),int((k-1)/2):int(w+(k-1)/2)] = img
    
    kernel = np.ones([k,k])
    
    kernel = kernel/sum(sum(kernel))
    
    for i in range(int((k-1)/2),h+int((k-1)/2)):
        for j in range(int((k-1)/2),w+int((k-1)/2)):     
            area_obs = padding_img[i-int((k-1)/2):i+int((k-1)/2)+1,j-int((k-1)/2):j+int((k-1)/2)+1].reshape(k,k)  
            output[i-int((k-1)/2)][j-int((k-1)/2)] = int(sum(sum(kernel*area_obs)))
#     print(output.shape)
    return output
```

```python
def adaptive_thresholding(img,k,sigma,threshold):
    h,w = img.shape
    
    output = np.zeros([h,w])
    
    padding_img = np.zeros([h+k-1,w+k-1],dtype='int64')
    padding_img[int((k-1)/2):int(h+(k-1)/2),int((k-1)/2):int(w+(k-1)/2)] = img
    
#     kernel = np.ones([k,k])/k**2
    kernel = np.zeros([k,k])
    
    center = int((k-1)/2)
    for i in range(k):
        for j in range(k):
            kernel[i][j] = np.exp(-(((i - center)**2 + (j-center)**2) / (2 * sigma * sigma)))
    
    kernel = kernel/sum(sum(kernel))
    
    for i in range(int((k-1)/2),h+int((k-1)/2)):
        for j in range(int((k-1)/2),w+int((k-1)/2)):     
            area_obs = padding_img[i-int((k-1)/2):i+int((k-1)/2)+1,j-int((k-1)/2):j+int((k-1)/2)+1].reshape(k,k)  
            output[i-int((k-1)/2)][j-int((k-1)/2)] = int(sum(sum(kernel*area_obs)))
    
    output[output<threshold] = 0
    
#     output[output>threshold] = 255
#     print(output.shape)
    return np.int64(output)
```

```python
edge_mask = adaptive_thresholding(smoothed,5,15,90)
plt.imshow(edge_mask,cmap='gray')
plt.show()
```
