```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# import plotly.express as px
import os
import time
IMAGES_FOLDER_PATH = "../images"
from scipy import signal
```

```python
f = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'lena.jpg'),0)
f = f[:256,:256]
plt.imshow(f, cmap='gray')
plt.show()

h = cv2.imread(os.path.join(IMAGES_FOLDER_PATH, 'camera.png'),0)
h = h[:256,:256]
plt.imshow(h, cmap='gray')
plt.show()

print("f.shape",f.shape)
print("h.shape",h.shape)
```

```python
def Recursive1DFFT(arr):
    n = arr.shape[0]
    if n==1:
        return arr

    if np.log2(n)-int(np.log2(n)) != 0.0:
        padding = 2**(int(np.log2(n))+1) - n
        arr = np.append(arr, np.zeros(padding))
        
    n = arr.shape[0]
    arr_even = arr[0:n:2]
    arr_odd = arr[1:n:2]
    
    y_even = Recursive1DFFT(arr_even)
    y_odd = Recursive1DFFT(arr_odd)
    
    E = np.array(np.exp((-2*np.pi*np.complex(0,1)*np.array(range(int(n/2))))/n))
    
    y= []
    y = y_even + E * y_odd
    y = np.append(y, y_even - E * y_odd)
    
    return y
```

```python
def Recursive2DFFT(arr):
    Y = np.zeros(arr.shape,dtype='complex')
    h,w = arr.shape
    
    for i in range(h):
        Y[i,:] = Recursive1DFFT(arr[int(i),:])[:w]
    Y_copy = np.copy(Y)
    for i in range(w):
        Y[:,i] = Recursive1DFFT(Y_copy[:,int(i)])[:h]
    
    return Y
```

```python
def IFFT2D(X):
    h,w = X.shape
    
    x = np.zeros(X.shape,dtype='complex')
    
    for i in range(h):
        x[i,:] = np.conj(Recursive1DFFT(np.conj(X[int(i),:]))[:w])/w
    x_copy = np.copy(x)
    for i in range(w):
        x[:,i] = np.conj(Recursive1DFFT(np.conj(x_copy[:,int(i)]))[:h])/h
        
    return x
```

```python
def convolve(f,g):
    h1,w1 = f.shape
    h2,w2 = g.shape
    
    aux = np.zeros(g.shape)
    g_dash = np.zeros(g.shape)
    for i in range(h2):
        aux[i,:] = g[h2-1-i,:]
    
    for i in range(w2):
        g_dash[:,i] = g[:,w2-1-i]
    print(g_dash.shape)
    output = np.zeros([h1,w1])
    
    padding_img = np.zeros([h1+h2-1,w1+w2-1])
    padding_img[int(h2/2):h1+int(h2/2),int(w2/2):w1+int(w2/2)] = f
    
    for i in range(int(h2/2),h1+int(h2/2)):
        for j in range(int(w2/2),w1+int(w2/2)):
            area_obs = padding_img[i-int(h2/2):i+int(h2/2)+1,j-int(w2/2):j+int(w2/2)+1]
#             print(g_dash.shape)
#             print(area_obs.shape)
#             print()
            val = sum(sum(g_dash*area_obs))
            output[i-int(h2/2),j-int(w2/2)] = val
    
    return np.int64(output)
```

## Part 1

```python
F = Recursive2DFFT(f)
H = Recursive2DFFT(h)
```

```python
ift_fh = IFFT2D(F*H)
```

```python
conv_fh = convolve(f,h)
```

```python
plt.figure(figsize =(12,10))
plt.subplot(121)
plt.imshow(np.abs(ift_fh),cmap='gray')
plt.title('IFT(F.*H)')
plt.subplot(122)
plt.imshow(np.abs(conv_fh),cmap='gray')
plt.title('Conv(f,h)')
plt.show()

# print(np.linalg.norm(np.abs(ift_fh)-np.abs(conv_fh)))
```

```python
F_np = np.fft.fft2(f)
H_np = np.fft.fft2(h)

plt.figure(figsize =(12,10))
plt.subplot(121)
plt.imshow(np.abs(np.fft.ifft2(F_np*H_np)),cmap='gray')
plt.title('Using inbuilt ifft(F.*H)')
plt.subplot(122)
plt.imshow(np.abs(ift_fh),cmap='gray')
plt.title('Using personal ifft(F.*H)')
plt.show()

res = signal.convolve2d(np.int64(f),np.int64(h))

plt.figure(figsize =(12,10))
plt.subplot(121)
plt.imshow(res,cmap='gray')
plt.title('Using inbuilt Convolve2D')
plt.subplot(122)
plt.imshow(np.abs(conv_fh),cmap='gray')
plt.title('Using Personal Conv2D')
plt.show()
```

```python
print("Mean square error : ",np.linalg.norm(np.abs(ift_fh)-np.abs(conv_fh))**2/np.product(conv_fh.shape))
```
