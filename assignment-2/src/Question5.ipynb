## Part 1

```python
def dft_2D_brute(img):
    dft = np.zeros(img.shape,dtype='complex')
    h,w = img.shape
    for i in range(h):
        for j in range(w):
            for ii in range(h):
                for jj in range(w):
                    # print(-2*np.pi*np.complex(0,((i*ii)/h + (j*jj)/w)))
                    dft[i,j] += img[ii,jj]*np.exp(-2*np.pi*np.complex(0,((i*ii)/h + (j*jj)/w)))
    return dft
```

```python
img = np.array([[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255]])
dft = dft_2D_brute(img)
```

```python
plt.figure(figsize =(12,10))
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(np.abs(np.fft.fftshift(dft)).astype('int64'),cmap='gray')
plt.title('2D DFT')
plt.show()
# plt.imshow(np.abs(np.fft.fftshift(np.fft.fft2(img))).astype('int64'),cmap='gray')
# plt.show()
```

## Part 2

```python
def Recursive1DFFT(arr):
    n = arr.shape[0]
    if n==1:
        return arr

    if np.log2(n)-int(np.log2(n)) != 0.0:
        padding = 2**(int(np.log2(n))+1) - n
        arr = np.append(arr, np.zeros(padding))
        
    n = arr.shape[0]
    arr_even = arr[0:n:2]
    arr_odd = arr[1:n:2]
    
    y_even = Recursive1DFFT(arr_even)
    y_odd = Recursive1DFFT(arr_odd)
    
    E = np.array(np.exp((-2*np.pi*np.complex(0,1)*np.array(range(int(n/2))))/n))
    
    y= []
    y = y_even + E * y_odd
    y = np.append(y, y_even - E * y_odd)
    
    return y
```

```python
def Recursive2DFFT(arr):
    Y = np.zeros(arr.shape,dtype='complex')
    h,w = arr.shape
    
    for i in range(h):
        Y[i,:] = Recursive1DFFT(arr[int(i),:])[:w]
    Y_copy = np.copy(Y)
    for i in range(w):
        Y[:,i] = Recursive1DFFT(Y_copy[:,int(i)])[:h]
    
    return Y
```

```python
img = np.array([[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255],[255,0,255,0,255,0,255,0],[0,255,0,255,0,255,0,255]])
dft = Recursive2DFFT(img)
```

## Part 3

```python
def IFFT2D(X):
    h,w = X.shape
    
    x = np.zeros(X.shape,dtype='complex')
    
    for i in range(h):
        x[i,:] = np.conj(Recursive1DFFT(np.conj(X[int(i),:]))[:w])/w
    x_copy = np.copy(x)
    for i in range(w):
        x[:,i] = np.conj(Recursive1DFFT(np.conj(x_copy[:,int(i)]))[:h])/h
        
    return x/h*w
```

```python
x = IFFT2D(dft)
```

```python
plt.figure(figsize =(12,10))
plt.subplot(131)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(132)
plt.imshow(np.abs(np.fft.fftshift(dft)).astype('int64'),cmap='gray')
plt.title('2D ABS DFT')
plt.subplot(133)
plt.imshow(np.real(x).astype('int64'),cmap='gray')
plt.title('IFFT')
plt.show()
```
